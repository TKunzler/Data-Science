```python
# Creating DataFrame with Dtype, Unique, and Null information
df_info = pd.DataFrame(df_past.dtypes, columns=['Dtype'])
df_info['Unique'] = df_past.nunique().values
df_info['Null'] = df_past.isnull().sum().values
display(df_info)


# Adicionando uma coluna para contar as ocorrncias
df_info['Count'] = 1 

# Contando quantos tipos de dados diferentes existem
count_by_dtype = df_info.groupby('Dtype').agg({'Count': 'sum'}).reset_index()
print('\n\nCounting dtypes')
display(count_by_dtype)
```

<div style="font-family: Montserrat; color:#02231C; background-color:#C7F6EC;">
    
<p style="background-color:#02231C; font-family:Montserrat; font-size:100%; text-align:left; border-radius:0px 10px 10px 0px; padding-left: 50px; width: 50%;"><strong><span style="color:#55D9C0">Is there any column that is 100% null?</span></strong></p>
    
- Teste

```python
# Contar o nmero de vezes que a string "na" aparece em cada coluna
na_string_counts = df_past.apply(lambda x: x.astype(str).str.lower().eq('na').sum())

# Ordenar os resultados do menor para o maior
na_string_counts_sorted = na_string_counts.sort_values()

# Exibir o resultado
print(na_string_counts_sorted)
```

<div style="font-family: Montserrat; color:#02231C; background-color:#C7F6EC;">
    
<p style="background-color:#02231C; font-family:Montserrat; font-size:100%; text-align:left; border-radius:0px 10px 10px 0px; padding-left: 50px; width: 50%;"><strong><span style="color:#55D9C0">NA Values</span></strong></p>
    
- **Due to the impossibility of obtaining further information about the null values, I will replace them with 0.**
    
    
- **Check ".describe()" after the transformation.**

```python
# Selecionar as colunas que queremos transformar, excluindo "class" e "aa_000"
cols_to_transform = df_present.columns.difference(['class', 'aa_000'])

# Substituir valores 'na' por '0' nas colunas selecionadas
df_present[cols_to_transform] = df_present[cols_to_transform].replace('na', '0')
df_past[cols_to_transform] = df_past[cols_to_transform].replace('na', '0')

# Substituir valores NaN por 0 nas colunas selecionadas
df_present[cols_to_transform] = df_present[cols_to_transform].fillna(0)
df_past[cols_to_transform] = df_past[cols_to_transform].fillna(0)

# Transformar as colunas selecionadas para float para lidar com possveis valores decimais
df_present[cols_to_transform] = df_present[cols_to_transform].astype(float)
df_past[cols_to_transform] = df_past[cols_to_transform].astype(float)

# Transformar as colunas selecionadas para int
df_present[cols_to_transform] = df_present[cols_to_transform].astype(int)
df_past[cols_to_transform] = df_past[cols_to_transform].astype(int)
```

```python
# Creating DataFrame with Dtype, Unique, and Null information
df_info = pd.DataFrame(df_past.dtypes, columns=['Dtype'])
df_info['Unique'] = df_past.nunique().values
df_info['Null'] = df_past.isnull().sum().values
display(df_info.head(7))


# Adicionando uma coluna para contar as ocorrncias
df_info['Count'] = 1 

# Contando quantos tipos de dados diferentes existem
count_by_dtype = df_info.groupby('Dtype').agg({'Count': 'sum'}).reset_index()
print('\n\nCounting dtypes')
display(count_by_dtype)
```

- **Through the "Report", I identified a large number of columns with high percentages of their values being zero.**
    
    
- **If a column has more than 95% of its values as zero, both in the complete dataframe and in cases where the class column is positive, it will be excluded.**   

```python
df_pos = df_past[df_past['class'] == 'pos'].copy()

over_95_list = []

# Funo para verificar porcentagem de zeros em uma coluna
def percent_zeros(series):
    return (series == 0).mean() > 0.92

# Verificao para df_past
for column in df_past.columns:
    if percent_zeros(df_past[column]) and percent_zeros(df_pos[column]):
        over_95_list.append(column)

# Mostrar as colunas que atendem ao critrio
print("Colunas com mais de 95% de zeros em ambos os dataframes: {}".format(len(over_95_list)))
print(over_95_list)
```

```python
# Excluir colunas com 
df_past = df_past.drop(columns=over_95_list, errors='ignore')

# Verificar o resultado
df_past.shape
```
